# Exploratory Data Analysis in Pandas

Importing libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Importing the dataset

```python
df = pd.read_csv(r"C:\Users\natee\OneDrive\Desktop\Pandas Datasets\world_population.csv")
df
```

Changing the display format of the floats throughout the dataframe

```python
pd.set_option('display.float_format', lambda x:'%.2f' % x)
df
```

Examining high level information about the dataset

```python
df.info()
```

```python
df.describe()
```

Checking where data is missing

```python
df.isnull().sum()
```

```python
df.nunique()
```

Exploring the top 10 and bottom 5 nations by population

```python
df.sort_values(by='2022 Population').head()
```

```python
df.sort_values(by='2022 Population', ascending=False).head(10)
```

Checking the correlations between different values in our dataset

```python
df.corr()
```

Using a heatmap to more quickly visualize correlations between values

```python
sns.heatmap(df.corr(), annot = True)

plt.rcParams['figure.figsize'] = (20,10) 

plt.show()
```

Grouping our dataframe by continent, sorted by population size

```python
df.groupby('Continent').mean().sort_values(by="2022 Population", ascending = False)
```

Checking out the data for the Oceania continent

```python
df[df['Continent'].str.contains('Oceania')].sort_values(by="2022 Population", ascending = False)
```

Slimming down and reorganizing the columns to create a line graph of population growth over time by continent

```python
df.columns
```

```python
df2 = df.groupby('Continent')['1970 Population','1980 Population','1990 Population','2000 Population',
                              '2010 Population','2015 Population', '2020 Population','2022 Population'].mean().sort_values(by="2022 Population", ascending = False)
df2
```

```python
df3 = df2.transpose()
df3
```

Viewing population change over time by continent

```python
df3.plot()
```

We can use boxplots to check for outliers

```python
df.boxplot(figsize=(20,10))
```
