# Merges, Joins, and Concatenations


```python
import pandas as pd
```

```python
df1 = pd.read_csv(r"C:\Users\natee\OneDrive\Desktop\Pandas Datasets\LOTR.csv")
df1
```

```python
df2 = pd.read_csv(r"C:\Users\natee\OneDrive\Desktop\Pandas Datasets\LOTR 2.csv")
df2
```

# Merges

Inner Merge

```python
df1.merge(df2, how='inner', on = ['FellowshipID', 'FirstName'])
```

Outer Merge

```python
df1.merge(df2, how='outer', on = ['FellowshipID', 'FirstName'])
```

Left Merge

```python
df1.merge(df2, how = 'left')
```

Right merge

```python
df1.merge(df2, how = 'right')
```

Cross Merge

```python
df1.merge(df2, how = 'cross')
```

# Joins

Outer Join

```python
df1.join(df2, on = 'FellowshipID', how = 'outer', lsuffix = '_Left', rsuffix = '_Right')
```

Joining using Fellowship ID on the index

```python
df4 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), lsuffix = '_Left', rsuffix = '_Right')
df4
```

```python
df4 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), lsuffix = '_Left', rsuffix = '_Right', how = 'outer')
df4
```

# Concatenating

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2], join = 'inner')
```

```python
pd.concat([df1,df2], join = 'outer')
```

```python
pd.concat([df1,df2], join = 'outer', axis = 1)
```
